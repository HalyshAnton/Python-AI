**Опис даних:**
* **Order**: Унікальний ідентифікатор для кожного запису в наборі даних.  
* **PID**: Номер ідентифікації власності.  
* **MS SubClass**: Тип житла (наприклад, односімейний будинок, дуплекс).  
* **MS Zoning**: Класифікація зонування власності (наприклад, житлова, комерційна).  
* **Lot Frontage**: Довжина вулиці, що прилягає до власності.  
* **Lot Area**: Загальна площа ділянки в квадратних футах.  
* **Street**: Тип доступу до дороги (наприклад, асфальтована, гравійна).  
* **Alley**: Тип доступу до провулка (якщо є).  
* **Lot Shape**: Загальна форма ділянки (наприклад, регулярна, нерегулярна).  
* **Land Contour**: Рівність або нахил землі (наприклад, рівна, горбиста).  
* **Utilities**: Тип доступних комунальних послуг (наприклад, всі публічні, лише електрика).  
* **Lot Config**: Конфігурація ділянки (наприклад, внутрішня ділянка, кутова ділянка).  
* **Land Slope**: Нахил землі (наприклад, легкий, помірний, сильний).  
* **Neighborhood**: Фізичні місця в межах міста Еймс.  
* **Condition 1**: Близькість до різних умов (наприклад, біля головної дороги, поруч із зеленим поясом).  
* **Condition 2**: Друга умова близькості (якщо є).  
* **Bldg Type**: Тип будівлі (наприклад, односімейний будинок, таунхаус).  
* **House Style**: Стиль будинку (наприклад, одноповерховий, двоповерховий).  
* **Overall Qual**: Загальна якість матеріалів та обробки (оцінюється від 1 до 10).  
* **Overall Cond**: Загальна оцінка стану (оцінюється від 1 до 10).  
* **Year Built**: Рік початкового будівництва будинку.  
* **Year Remod/Add**: Рік ремонту або добудов.  
* **Roof Style**: Тип даху (наприклад, двосхилий, шатровий).  
* **Roof Matl**: Матеріал даху (наприклад, асфальт, метал).  
* **Exterior 1st**: Зовнішнє покриття будинку (наприклад, вініл, цегла).  
* **Exterior 2nd**: Друге зовнішнє покриття (якщо є).  
* **Mas Vnr Type**: Тип кам'яного облицювання (наприклад, цегла, камінь).  
* **Mas Vnr Area**: Площа кам'яного облицювання в квадратних футах.  
* **Exter Qual**: Якість зовнішньої обробки (наприклад, відмінна, добра).  
* **Exter Cond**: Стан зовнішньої обробки (наприклад, відмінний, добрий).  
* **Foundation**: Тип фундаменту (наприклад, монолітний бетон, плита).  
* **Bsmt Qual**: Висота підвалу (наприклад, відмінна, добра).  
* **Bsmt Cond**: Стан підвалу (наприклад, відмінний, добрий).  
* **Bsmt Exposure**: Вихід або відкритість до зовнішнього середовища (наприклад, ні, мінімальна).  
* **BsmtFin Type 1**: Тип обробленої зони в підвалі (наприклад, кімната відпочинку, спальня).  
* **BsmtFin SF 1**: Квадратні фути обробленої зони в підвалі (Тип 1).  
* **BsmtFin Type 2**: Другий тип обробленої зони в підвалі (якщо є).  
* **BsmtFin SF 2**: Квадратні фути обробленої зони в підвалі (Тип 2).  
* **Bsmt Unf SF**: Квадратні фути необробленого підвалу.  
* **Total Bsmt SF**: Загальна площа підвалу (оброблена і необроблена).  
* **Heating**: Тип опалення (наприклад, газ, електрика).  
* **Heating QC**: Якість опалення (наприклад, відмінна, добра).  
* **Central Air**: Наявність центрального кондиціонера (Так/Ні).  
* **Electrical**: Тип електросистеми (наприклад, стандартна, на основі ручних вимикачів).  
* **1st Flr SF**: Квадратні фути першого поверху.  
* **2nd Flr SF**: Квадратні фути другого поверху.  
* **Low Qual Fin SF**: Квадратні фути низькоякісної обробленої площі.  
* **Gr Liv Area**: Площа житлового простору над землею в квадратних футах.  
* **Bsmt Full Bath**: Кількість повних ванних кімнат у підвалі.  
* **Bsmt Half Bath**: Кількість напівванних кімнат у підвалі.  
* **Full Bath**: Кількість повних ванних кімнат у будинку.  
* **Half Bath**: Кількість напівванних кімнат у будинку.  
* **Bedroom AbvGr**: Кількість спалень над землею.  
* **Kitchen AbvGr**: Кількість кухонь над землею.  
* **Kitchen Qual**: Якість кухні (наприклад, відмінна, добра).  
* **TotRms AbvGrd**: Загальна кількість кімнат над землею.  
* **Functional**: Оцінка функціональності будинку (наприклад, повністю функціональний, незначні недоліки).  
* **Fireplaces**: Кількість камінів.  
* **Fireplace Qu**: Якість каміна (наприклад, відмінна, добра).  
* **Garage Type**: Тип гаража (наприклад, прибудований, окремо стоячий).  
* **Garage Yr Blt**: Рік будівництва гаража.  
* **Garage Finish**: Внутрішнє оздоблення гаража (наприклад, необроблений, готовий).  
* **Garage Cars**: Кількість автомобілів, які може вмістити гараж.  
* **Garage Area**: Розмір гаража в квадратних футах.  
* **Garage Qual**: Якість гаража (наприклад, відмінна, добра).  
* **Garage Cond**: Стан гаража (наприклад, відмінний, добрий).  
* **Paved Drive**: Тип під'їзної дороги (наприклад, асфальтована, гравійна).  
* **Wood Deck SF**: Площа дерев'яної тераси в квадратних футах.  
* **Open Porch SF**: Площа відкритого ганку в квадратних футах.  
* **Enclosed Porch**: Площа закритого ганку в квадратних футах.  
* **3Ssn Porch**: Площа трисезонного ганку в квадратних футах.  
* **Screen Porch**: Площа ганку з москітною сіткою в квадратних футах.  
* **Pool Area**: Розмір площі басейну в квадратних футах.  
* **Pool QC**: Якість басейну (наприклад, відмінна, добра).  
* **Fence**: Тип огорожі (наприклад, GdWo, MnPrv).  
* **Misc Feature**: Додаткова характеристика (наприклад, сарай, тенісний корт).  
* **Misc Val**: Вартість додаткових характеристик.  
* **Mo Sold**: Місяць продажу нерухомості.  
* **Yr Sold**: Рік продажу нерухомості.  
* **Sale Type**: Тип продажу (наприклад, новий, частковий).  
* **Sale Condition**: Умова продажу (наприклад, нормальна, ненормальна).
